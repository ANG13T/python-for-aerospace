## What is Matplotlib? ðŸ“ˆ
Matplotlib is a library for creating visuals with Python. This could include line charts, graphs, various plot types, and more!
You can learn more about the capabilities of Matplotlib on their website: [https://matplotlib.org/](https://matplotlib.org/)

Let's begin by importing the matplotlib library under the alias `plt`

NOTE: we are using the pyplot module which provides basic functions for graphing

```python
import matplotlib.pyplot as plt
```

Let's being by creating a simple line plot using matplotlib!

```python
# Create data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Plot data
plt.plot(x, y)

# Show plot
plt.show()
```

Let's spruce up the visuals of our plot by adding some color and chaning the line appearance!

```python
# Customize appearance
plt.plot(x, y, color='green', linestyle='--', marker='o', markersize=5)

# Show plot
plt.show()
```

Now, let's finish up our plot by adding a title alongside some text for each of the axes.

```python
# Add labels and a title
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('My First Line Plot!')

# Show plot
plt.show()
```

## More Visualizations with Matplotlib ðŸŽ¨

You can make more complex visuals using matplotlib using other methods and by utlizing other libraries in tandem such as `math` and `numpy`

```python
import math
import numpy as np
```

The following code snippet visualizes the `sin` function using the `math` library.
NOTE: we use the `linspace` method using numpy to create evenly spaced values between 0 and 2 pi.

```python
x = np.linspace(0, 2*math.pi, 100)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Plot of sin(x)')
plt.grid(True)
plt.show()
```

```python
x = np.linspace(0, 2*math.pi, 100)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Customized Plot of sin(x)')
plt.legend()
plt.grid(True)
plt.show()
```

You can also plot multiple mathematical functions on the same axis!

```python
x = np.linspace(0, 2*math.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.xlabel('x')
plt.ylabel('y')
```

Matplotlib can also be used to plot three-dimensional surfaces using the `mpl_toolkits.mplot3d` module

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot of sin(sqrt(X^2 + Y^2))')

plt.show()
```
