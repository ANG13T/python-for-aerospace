## What is a Differential Equation?
A differential equation (DE) is a mathematical equation that relates one or more unknown functions to their derivatives. In other words, it describes how a function or a set of functions change with respect to one or more independent variables. 

![https://www.mathsisfun.com/calculus/images/diff-eq-1.svg](https://www.mathsisfun.com/calculus/images/diff-eq-1.svg)

[Image Source](https://www.mathsisfun.com/calculus/differential-equations.html)

DE's form the backbone behind a number of aerospace applications such as flight dynamics, orbital mechanics, aerodynamics, and much more.

For example, the Rocket Equation (also known as the Tsiolkovsky rocket equation) can be modeled using a differential equation to describe the rate of change of a rocket's velocity with respect to time as a function of the mass flow rate of propellant expelled from the rocket and the velocity of the exhaust gases relative to the rocket.

## Differential Equations using SciPy
Let's use SciPy to solve an Ordinary Differential Equation (ODE), which is a differential equation involving a single independent variable and its derivatives. The ODE we will be solving is as follows:

$$
\frac{dy}{dt} = -yt + 5
$$

```python
import numpy as np 
from scipy.integrate import odeint 
import matplotlib.pyplot as plt 
```

```python
def differential_equation(y,t): 
    dydt = -y * t + 5
    return dydt
```

```python
# initial condition 
y0 = 1
  
# values of time 
t = np.linspace(0,5) 
  
# Solving ODE 
y = odeint(differential_equation, y0, t) 
```

```python
plt.plot(t,y) 
plt.xlabel("Time") 
plt.ylabel("Y") 
plt.title("Visualizing dy/dt ODE")
plt.show()
```

## Vertical Acceleration ODE 
The vertical motion of a rocket can be defined by the following equation.
$$
a = (thrust - m*g) / m
$$
Using `scipy` we can model the motio of the rocket using a ODE and then solve it + plot it on a graph!

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
g = 9.81  # Acceleration due to gravity (m/s^2)
m = 1000  # Mass of the rocket (kg)
thrust = 20000  # Thrust of the rocket (N)
```

```python
def rocket_motion_ode(t, state):
    # Extract state variables
    x, v = state

    # Compute acceleration (assuming vertical motion)
    a = (thrust - m*g) / m

    return [v, a]
```

```python
x0 = 0  # Initial position (m)
v0 = 0  # Initial velocity (m/s)
state0 = [x0, v0]
t_span = (0, 100)
```

```python
sol = solve_ivp(rocket_motion_ode, t_span, state0, t_eval=np.linspace(t_span[0], t_span[1], 1000))
```

```python
plt.plot(sol.t, sol.y[0])
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.title('Rocket Trajectory')
plt.grid(True)
plt.show()
```
