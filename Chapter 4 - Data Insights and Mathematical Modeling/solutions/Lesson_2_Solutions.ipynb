## Exercise 1: Modeling a Differential Equation using SciPy

#### Objectives
- Create a differential equation (ODE) that models the following equation:
$$
\frac{dy}{dt} = y^2t + 8t
$$
- Set the intial y condition to be `1` and the time range to be `1 - 2`
- Visualize the ODE using `matplotlib`

```python
import numpy as np 
from scipy.integrate import odeint 
import matplotlib.pyplot as plt 
```

```python
def differential_equation(y, t):
    dydt = (y ** 2)  * t + (8 * t)
    return dydt
```

```python
# initial condition 
y0 = 1
  
# values of time 
t = np.linspace(1,2) 
  
# Solving ODE 
y = odeint(differential_equation, y0, t) 
```

```python
plt.plot(t,y) 
plt.xlabel("Time") 
plt.ylabel("Y") 
plt.title("Visualizing dy/dt ODE")
plt.show()
```

## Exercise 2: Modeling Rocket Horizontal Motion ODE

#### Objectives:
- Create a function for the following differential equation describing horizontal motion of a rocket:
$$
\frac{dv}{dt} = \frac{T - D}{m}
$$
- Set the initial horizontal velocity to be 0
- Set the time span to be 0 - 100 s
- Plot the ODE solution using `matplotlib`

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
# Constants
m = 1000  # Mass of the rocket (kg)
T = 20000  # Thrust of the rocket (N)
```

```python
def rocket_motion_ode(t, state):    
    # Extract state variable (horizontal velocity)
    v = state[0]
    D = 500*t  # Drag force as a function of time (N)
    # Compute acceleration (Newton's second law)
    a = (T - D) / m

    return [a]
```

```python
v0 = 0  # Initial horizontal velocity (m/s)
state0 = [v0]
# Time span
t_span = (0, 100)
```

```python
sol = solve_ivp(rocket_motion_ode, t_span, state0, t_eval=np.linspace(t_span[0], t_span[1], 1000))
```

```python
plt.plot(sol.t, sol.y[0])
plt.xlabel('Time (s)')
plt.ylabel('Horizontal Velocity (m/s)')
plt.title('Rocket Horizontal Motion')
plt.grid(True)
plt.show()
```
