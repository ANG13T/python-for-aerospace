# Lesson 3: Partial Differential Equations

## What are Partial Differential Equations (PDEs)?
Unlike ordinary differential equations (ODEs), which involve only one independent variable, PDEs involve functions of two or more independent variables and describe how these functions vary with respect to each independent variable. An example of a PDE is as follows:

$$
\frac{d^2u}{dx^2}+\frac{d^2u}{dy^2} = 0
$$

Where `u` is the dependent variable and `x` and `y` are independent variables.

Within a physical context, partial differential equations are used to model situations where more than one independent variable affects the behavior of a system. These situations often involve phenomena that vary in space and time, or across multiple spatial dimensions.

Partial differential equations open a wide range of doors in terms of their relevant application within the space sector. Some examples of applications include:
- Orbital Mechanics
- Astrodynamics and Mission Design
- Spacecraft Thermal Analysis

## Solving PDEs using SciPy
We can leverage the tools provided by SciPy to solve PDEs effectively. For example, consider the following PDE used to describe how  the temperature u(x,t) of a one-dimensional rod varies over time as heat diffuses through the material.

$$
\frac{du}{dt} = α\frac{d^2u}{dx^2}
$$

Where the variables are defined as follows:
- `u(x,t)` is the temperature at position x and time t
- `α` thermal diffusivity of the material
- `du/dt` rate of change of temperature with respect to time
- `d^2u/dx^2` spatial curvature of the temperature profile

![https://upload.wikimedia.org/wikipedia/commons/thumb/9/97/Temp_Rod_homobc.svg/300px-Temp_Rod_homobc.svg.png](https://upload.wikimedia.org/wikipedia/commons/thumb/9/97/Temp_Rod_homobc.svg/300px-Temp_Rod_homobc.svg.png)

[Image Source](https://en.wikipedia.org/wiki/Heat_equation)

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
L = 1  # Length of the rod
alpha = 0.01  # Thermal diffusivity
T_initial = lambda x: np.sin(np.pi * x)  # Initial temperature distribution
T_left_boundary = 0  # Temperature at left boundary
T_right_boundary = 0  # Temperature at right boundary
```

```python
def heat_equation_pde(t, T):
    dTdx2 = np.gradient(np.gradient(T))  # Second spatial derivative
    return alpha * dTdx2
```

```python
x = np.linspace(0, L, 100)
T0 = T_initial(x)
sol = solve_ivp(heat_equation_pde, [0, 1], T0, t_eval=np.linspace(0, 1, 100))
```

```python
plt.figure(figsize=(10, 6))
for i in range(len(sol.t)):
    plt.plot(x, sol.y[:, i], label=f't = {sol.t[i]:.2f}')
plt.xlabel('Position (x)')
plt.ylabel('Temperature (T)')
plt.title('Temperature Distribution Over Time (1D Heat Equation)')
plt.grid(True)
plt.show()
```

## Exercise 1: Laplace Equation PDE
The Laplace Equation is used in a wide range of aerospace context, particularly in problems involving steady-state heat conduction, electrostatics, fluid flow, and potential flow around aerodynamic bodies. It is defined by the following PDE:

$$
∇^2u = \frac{d^2u}{dx^2} + \frac{d^2u}{dy^2} = 0
$$
