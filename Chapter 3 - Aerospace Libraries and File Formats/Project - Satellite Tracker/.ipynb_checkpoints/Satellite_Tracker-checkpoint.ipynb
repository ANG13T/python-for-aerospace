## Create a Ground Track Visualization using skyfield

### Step 1: Intialize Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from skyfield.api import EarthSatellite, load
from datetime import datetime, timedelta, timezone
```

### Step 2: Define TLE Data

```python
line_one = "1 25544U 98067A   21091.47054420  .00001472  00000-0  37181-4 0  9995"
line_two = "2 25544  51.6450  71.1210 0002893  77.8770  20.6756 15.48978768396849"
name = "ISS"
satellite = EarthSatellite(line_one, line_two, name)
```

### Step 3: Define Time Range

```python
# Create a timescale
ts = load.timescale()

t_start = datetime.utcnow().replace(tzinfo=timezone.utc)
t_end = t_start + timedelta(days=1)
t = np.linspace(0, 1, num=10)
times = t_start + t * (t_end - t_start)
```

### Step 4: Extract the Latitude and Longitude Coordinates

```python
lats, lons = [], []
for time in times:
    subpoint = satellite.at(ts.utc(time)).subpoint()
    lats.append(subpoint.latitude.degrees)
    lons.append(subpoint.longitude.degrees)
```

### Step 5: Visualize the Ground Track using matplotlib

```python
plt.figure(figsize=(10, 7))
plt.plot(lons, lats, marker='o', markersize=3, linestyle='-', color='blue')
plt.xlabel('Longitude (degrees)')
plt.ylabel('Latitude (degrees)')
plt.title('Ground Track of Satellite')
plt.grid(True)
plt.show()
```
