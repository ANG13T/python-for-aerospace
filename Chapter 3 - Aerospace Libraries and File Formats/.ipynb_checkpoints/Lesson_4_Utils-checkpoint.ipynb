## What is Photutils?
Photutils is a library written in Python for performing photometry (measurement of the intensity of light emitted or reflected by astronomical objects). It provides a range of tools for detecting and measuring sources in astronomical image and other tasks. You can learn more about it here:
[https://photutils.readthedocs.io/en/stable/](https://photutils.readthedocs.io/en/stable/)

## Source Detection with Photutils

```python
from astropy.io import fits
import astropy.units as u
from astropy.nddata import CCDData
from astropy.stats import sigma_clipped_stats, SigmaClip
from astropy.visualization import ImageNormalize, LogStretch
import matplotlib.pyplot as plt
from matplotlib.ticker import LogLocator
import numpy as np
from photutils.background import Background2D, MeanBackground

# Retrieves the FITS file we will be using for our example
url = 'https://archive.stsci.edu/pub/hlsp/xdf/hlsp_xdf_hst_acswfc-60mas_hudf_f435w_v1_sci.fits'
with fits.open(url) as hdulist:
    hdulist.info()
    data = hdulist[0].data
    header = hdulist[0].header
```

```python
unit = u.electron/ u.s
mask = data == 0
xdf_image = CCDData(data, unit=unit, meta=header, mask=mask)
# Set up the figure with subplots
fig, ax1 = plt.subplots(1, 1, figsize=(8, 8))


# Set up the normalization and colormap
norm_image = ImageNormalize(vmin=1e-4, vmax=5e-2, stretch=LogStretch(), clip=False)
cmap = plt.get_cmap('viridis')
cmap.set_over(cmap.colors[-1])
cmap.set_under(cmap.colors[0])
cmap.set_bad('white') # Show masked data as white
xdf_image_clipped = np.clip(xdf_image, 1e-4, None) # clip to plot with logarithmic stretch

# Plot the data
fitsplot = ax1.imshow(np.ma.masked_where(xdf_image.mask, xdf_image_clipped), 
                      norm=norm_image, cmap=cmap)

# Define the colorbar and fix the labels
cbar = plt.colorbar(fitsplot, fraction=0.046, pad=0.04, ticks=LogLocator(subs=range(10)))
labels = ['$10^{-4}$'] + [''] * 8 + ['$10^{-3}$'] + [''] * 8 + ['$10^{-2}$']
cbar.ax.set_yticklabels(labels)

# Define labels
cbar.set_label(r'Flux Count Rate ({})'.format(xdf_image.unit.to_string('latex')), 
               rotation=270, labelpad=30)
ax1.set_xlabel('X (pixels)')
ax1.set_ylabel('Y (pixels)')
```

## What is Specutils
Specutils is a Python library for spectroscopic data analysis in astronomy. It provides a set of tools and utilities for working with spectroscopic data (measurements of light intensity as a function of wavelength or frequency). You can learn more about it here:
[https://specutils.readthedocs.io/en/stable/](https://specutils.readthedocs.io/en/stable/).

### What is Spectroscopy?
Spectroscopy is the study of how different materials interact with light across various wavelengths. By analyzing how light is transmitted, reflected, and absorbed against a specific surface, we can obtain general characteristics of the material such as composition and structure.

Spectroscopic images are visualized using spectra representing the intensity of light at different wavelengths across the image. These spectra are typically displayed as plots or graphs, where the x-axis represents the wavelength or frequency of light, and the y-axis represents the intensity or amplitude of the signal.

## Fitting and Visualization using Specutils
We can use `specutils` to plot spectra and create a fitting model to best represent the spectra using a function. 
The following code will initialize a random spectra using the predefined `generate_random_spectra` function and then creates a simple fitting model using the `Gaussian1D` model. Once the data is model fitted, it is then visualized using `matplotlib`

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.modeling import models
from astropy import units as u
from specutils.spectra import Spectrum1D
from specutils.fitting import fit_lines
```

```python
# Preconfigured code for generating a random spectra
np.random.seed(0)
x = np.linspace(0., 10., 200)
y = 3 * np.exp(-0.5 * (x- 6.3)**2 / 0.8**2)
y += np.random.normal(0., 0.2, x.shape)
spectra = Spectrum1D(flux=y*u.Jy, spectral_axis=x*u.um)
```

```python
# Calculate fitted flux values and fit the spectrum using the Gaussian1D model
gaussian = models.Gaussian1D(amplitude=3.*u.Jy, mean=6.1*u.um, stddev=1.*u.um)
gaussian_fitted = fit_lines(spectra, gaussian)
y_gaussian_fit = gaussian_fitted(x*u.um)
```

```python
# Plot the original spectra and the fitted model
plt.plot(x, y, label="Original spectrum")
plt.plot(x, y_gaussian_fit, label="Fitted model")
plt.title('Fitting Spectra using specutils')
plt.grid(True)
plt.legend()
```

### What is Continuum Fitting?
Continuum fitting is a technique used in spectroscopy to model and remove the continuum (background signal) from a spectra. 
This is helpful for removing "background noise" from a spectral data set in order to focus on the relevant spectroscopic details.
